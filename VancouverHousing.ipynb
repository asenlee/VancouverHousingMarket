```python
# import necessary packages
import pandas as pd
import numpy as np
import matplotlib.pylab as pylab
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv file
van_housing = pd.read_csv("van_housing_Jun_13_21_clean.csv")
van_housing.head()
```

## Exploratory Data Analysis

```python
van_housing.count()
```

After removing the duplicated URLs, there are 120 unique posts. We can observe that some posts are missing `num_bedrooms` and `sqft`.

```python
# scatter plot of Price vs Sqft
params = {'legend.fontsize': 'x-large', 
        'figure.figsize': (15,5),
        'axes.labelsize': 'x-large',
        'axes.titlesize': 'x-large',
        'xtick.labelsize': 'x-large',
        'ytick.labelsize': 'x-large'}
pylab.rcParams.update(params)

plt.figure(figsize =(12,8))
sns.scatterplot(x='price', y='sqft', hue='num_bedrooms', palette='summer', x_jitter=True, y_jitter=True, s=125, data=van_housing.dropna())
plt.legend(fontsize=12)
plt.xlabel("Price", fontsize=15)
plt.ylabel("Square Footage", fontsize = 15)
plt.title("Price vs Square Footage Colored by Number of Bedrooms", fontsize=15)
```
